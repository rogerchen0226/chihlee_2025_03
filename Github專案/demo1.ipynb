- 這裡是一個建立*正弦波*的圖表範




```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和取樣點
x = np.linspace(0, 2 * np.pi, 100)

# 計算 y 值 (sin(x))
y = np.sin(x)

# 繪製圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('sin wave')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 顯示圖表
plt.show()
```

- 同時顯示 sin cos 的正弦波和餘弦波

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍和取樣點
x = np.linspace(0, 2 * np.pi, 100)

# 計算 y 值 (sin(x) 和 cos(x))
y_sin = np.sin(x)
y_cos = np.cos(x)

# 繪製圖表
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')

# 設定圖表標題和軸標籤
plt.title('sin and cos waves')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖例
plt.legend()

# 顯示圖表
plt.show()
```
